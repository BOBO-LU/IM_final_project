```python
# opview 資料（聲量）
path ='D:\\IM_final_project-main\\Stephanie20220107\\trend in list merged\\'
emotion_path = 'D:\\IM_final_project-main\\Stephanie20220107\\emotion in list merged\\'
file_path = 'D:\\IM_final_project-main\\temp\\'
for i in os.listdir('./Stephanie20220107/trend in list merged/'):
    name =  i.split('_trend')[0]
    op_view_df = pd.read_csv(path + i )
    op_view_df['temp'] = op_view_df['聲量'].rolling(7).mean().shift(1)
    op_view_df['Youtuber'] = name
    op_view_df['前一天聲量'] = op_view_df['聲量'].rolling(1).sum().shift(1)
    op_view_df['前二天聲量'] = op_view_df['聲量'].rolling(1).sum().shift(2)
    op_view_df['後一天聲量'] = op_view_df['聲量'].rolling(1).sum().shift(-1)
    op_view_df['後二天聲量'] = op_view_df['聲量'].rolling(1).sum().shift(-2)
    op_view_df['當天聲量變化量'] = op_view_df['聲量'] - op_view_df['前一天聲量']
    op_view_df['隔天聲量變化量'] = op_view_df['後一天聲量'] - op_view_df['聲量']
    op_view_df['前一後一聲量變化量'] = op_view_df['後一天聲量'] - op_view_df['前一天聲量']
    op_view_df['前二後二聲量變化量'] = op_view_df['後二天聲量'] - op_view_df['前二天聲量']

    MULTIPLY = 3
    event_date_list = op_view_df[ op_view_df['聲量'] > MULTIPLY * op_view_df['temp']]['期間'].to_list()
    main_df = op_view_df[ op_view_df['聲量'] > MULTIPLY * op_view_df['temp']].drop(['temp'], axis = 1)


    #emotion
    
    emotion_df = pd.read_csv(emotion_path + name + '_emotion_2019to2021.csv')


    emotion_df['正面情緒(前一天)'] = emotion_df['正面情緒'].rolling(1).sum().shift(1)
    emotion_df['正面情緒(前二天)'] = emotion_df['正面情緒'].rolling(1).sum().shift(2)
    emotion_df['正面情緒(後一天)'] = emotion_df['正面情緒'].rolling(1).sum().shift(-1)
    emotion_df['正面情緒(後二天)'] = emotion_df['正面情緒'].rolling(1).sum().shift(-2)

    emotion_df['負面情緒(前一天)'] = emotion_df['負面情緒'].rolling(1).sum().shift(1)
    emotion_df['負面情緒(前二天)'] = emotion_df['負面情緒'].rolling(1).sum().shift(2)
    emotion_df['負面情緒(後一天)'] = emotion_df['負面情緒'].rolling(1).sum().shift(-1)
    emotion_df['負面情緒(後二天)'] = emotion_df['負面情緒'].rolling(1).sum().shift(-2)

    emotion_df['當天正負情緒比'] = emotion_df['正面情緒'] / emotion_df['負面情緒']
    emotion_df['前一天正負情緒比'] = emotion_df['當天正負情緒比'].rolling(1).sum().shift(1)
    emotion_df['前二天正負情緒比'] = emotion_df['當天正負情緒比'].rolling(1).sum().shift(2)
    emotion_df['後一天正負情緒比'] = emotion_df['當天正負情緒比'].rolling(1).sum().shift(-1)
    emotion_df['後二天正負情緒比'] = emotion_df['當天正負情緒比'].rolling(1).sum().shift(-2)
    emotion_df['當天情緒比變化量'] = emotion_df['當天正負情緒比'] - emotion_df['前一天正負情緒比']
    emotion_df['隔天情緒比變化量'] = emotion_df['後一天正負情緒比'] - emotion_df['當天正負情緒比'] 
    emotion_df['前一後一情緒變化量'] = emotion_df['後一天正負情緒比'] - emotion_df['前一天正負情緒比'] 
    emotion_df['前二後二情緒變化量'] = emotion_df['後二天正負情緒比'] - emotion_df['前二天正負情緒比'] 

    emotion_event_df = emotion_df[emotion_df['期間'].isin(event_date_list)]
    emotion_event_df.drop('總聲量', axis = 1, inplace=True)

    final_df = emotion_event_df.merge(main_df, on = '期間')
    
    final_df.to_csv(name + '.csv', index=False)
```

```python
#news

file_path = 'D:\\IM_final_project-main\\temp\\'
news_path = 'D:\\IM_final_project-main\\data\\'
for i in (os.listdir(file_path)):
    name = i.split('.csv')[0]
#     print(name)
    event_df = pd.read_csv(file_path + i)
    event_date_list = event_df['期間'].to_list()
    temp_df = pd.DataFrame(columns = ['Source','Title','Link','Date','Summary','Text','Like'])
    for j in os.listdir(news_path + name):
        if (j != '.ipynb_checkpoints'):
            
            read_df = pd.read_csv(news_path + name + '\\' + j )
            read_df['Date'] =  pd.to_datetime(read_df['Date']).dt.strftime('%Y/%m/%d')
            temp_df = temp_df.append(read_df, ignore_index = True)
#             print(name,j, read_df['Source'])
    temp_df = temp_df.drop_duplicates(subset='Date', keep="first")
    
    event_news = pd.DataFrame(columns=['期間','Dcard','Dcard內文','PTT','PTT內文','apple_news', 'apple_news內文','ettoday', 'ettoday內文','Mobile01' ,'Mobile01內文'])

    temp = temp_df[temp_df['Date'].isin(event_date_list)]
    count = 0
    for i in event_date_list:
        a = temp_df[temp_df['Date'] == i]
        dcard = a[a['Source'] == 'Dcard']['Title'].values
        dcardc = a[a['Source'] == 'Dcard']['Text'].values        
        ptt = a[a['Source'] == 'PTT']['Title'].values
        pttc = a[a['Source'] == 'PTT']['Text'].values
        apple = a[a['Source'] == 'apple_news']['Title'].values
        applec = a[a['Source'] == 'apple_news']['Text'].values
        et = a[a['Source'] == 'ettoday']['Title'].values
        etc = a[a['Source'] == 'ettoday']['Text'].values
        mb = a[a['Source'] == 'Mobile01']['Title'].values
        mbc = a[a['Source'] == 'Mobile01']['Text'].values
        event_news.loc[count] = [i, dcard, dcardc, ptt,pttc, apple,applec, et,etc,mb,mbc]
    #     event_news.loc[count] = []
        count += 1
    event_news
    final_df = event_df.merge(event_news, on = '期間')
    final_df.to_csv(name + '_final.csv', index = False)
```
