```python
import pandas as pd
import os
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
col = pd.read_csv('./final/Dinter_final.csv')
```

```python
merge_final = pd.DataFrame(columns = list(col.columns))
for i in os.listdir('./final'):
    data = pd.read_csv('./final/'+ i)
    name = i.split('_final')[0]
    data['Youtuber'] = name
    merge_final = merge_final.append(data)
    print(i)
merge_final
```

```python
merge_final['Youtuber'].value_counts()
```

```python
bobo = pd.read_csv('ytr_data.csv').drop('Unnamed: 0', axis = 1)
```

```python
merged_with_bobo = merge_final.merge(bobo, on ='Youtuber', how='left')
merged_with_bobo
```

```python
merged_with_bobo.to_csv('merged_with_bobo.csv', index=False)
```
